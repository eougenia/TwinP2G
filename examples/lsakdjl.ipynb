```python
import pandas as pd
```

```python
# Read the CSV file
dfhv = pd.read_csv(r"C:\test\csvfolder\HVCUSTCONS_20201001_20201031 (1).csv")
#20201025,25,724.9618378,37.28140312

# Convert the date and hour columns to strings
dfhv['date'] = dfhv['date'].astype(str)
dfhv['hour'] = dfhv['hour'].astype(str)

# Pad the hour values with leading zeros if necessary
dfhv['hour'] = dfhv['hour'].str.zfill(2)

# Adjust hour '24' to '00' and increment the date by one day
dfhv.loc[dfhv['hour'] == '24', 'hour'] = '00'
dfhv['date'] = pd.to_datetime(dfhv['date'], format='%Y%m%d')
dfhv.loc[dfhv['hour'] == '00', 'date'] += pd.DateOffset(days=1)
dfhv['date'] = dfhv['date'].dt.strftime('%Y-%m-%d')

# Merge the date and hour columns
dfhv['datetime'] = dfhv['date'] + ' ' + dfhv['hour'] + ':00:00'

# Optionally, convert the 'datetime' column to a datetime data type
dfhv['datetime'] = pd.to_datetime(dfhv['datetime'])
dfhv.set_index(['datetime'],inplace=True)
dfhv

```
