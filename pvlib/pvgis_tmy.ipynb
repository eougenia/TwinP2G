```python
import pvlib
import pandas as pd
```

```python
latitude=37.983810 
longitude=23.727539
surface_tilt=0 
surface_azimuth=180
startyear=None
endyear=None
```

```python
tmy=pvlib.iotools.get_pvgis_tmy(latitude, 
                            longitude, 
                            outputformat='csv', 
                            usehorizon=True, 
                            userhorizon=None, 
                            startyear=startyear, 
                            endyear=endyear, 
                            map_variables=True, 
                            url='https://re.jrc.ec.europa.eu/api/v5_2/', 
                            timeout=30)[0]
```

```python
tmy.columns
```

```python
def extend_tmy_data(df, num_years):
    extended_dfs = []

    for year_offset in range(num_years):
        extended_dfs.append(df.copy())

    extended_data = pd.concat(extended_dfs, ignore_index=True)
    return extended_data
```

```python
num_years = 5
extended_tmy_data = extend_tmy_data(tmy, num_years)
extended_tmy_data.index=pd.date_range(start='2021-01-01 00:00', end='2025-12-30 23:00', freq='h')
extended_tmy_data
```

```python
extended_tmy_data.to_csv('pvgis_tmy.csv', index=True)
```
