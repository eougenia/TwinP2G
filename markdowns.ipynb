# Power-to Gas Συστήματα
Τα συστήματα Power-to-Gas (P2G) αναπτύσσονται ως μια καινοτόμος τεχνολογία αποθήκευσης ενέργειας και μετατροπής περιβαλλοντικά φιλικών πηγών ενέργειας σε αέρια καύσιμα. Τα συστήματα P2G χρησιμοποιούν περίσσεια παραγωγή ενέργειας από ανανεώσιμες πηγές, όπως η αιολική και η ηλιακή ενέργεια, για την παραγωγή υδρογόνου ή μεθανίου. Τα καύσιμα αυτά μπορούν να αποθηκευτούν και να χρησιμοποιηθούν αργότερα για την παραγωγή ηλεκτρισμού ή ως καύσιμα για οχήματα.

## Δεδομένα Παραγωγής και Κατανάλωσης
Για τα δεδομένα παραγωγής και κατανάλωσης χρησιμοποιήθηκε το αρχείο "time_series_60min_singleindex.csv" από τον ENTSO-E με διάρκεια 5 ετών και 9 μηνών ανά μια ώρα για την περίοδο 01-01-2015 - 30-09-2020.\
Το αρχείο περιλαμβάνει την συνολική παραγωγή φωτοβολταικών της χώρας καθώς και την συνολική ζήτηση ενέργειας.\
Για τον λόγο αυτό τροποποιήθηκε ώστε να αντικατοπτρίζει μία περιοχή της Ελλάδας με περιπου 5000 κάτοικους, και επομένως διαιρέθηκε η συνολική κατανάλωση με 2000 και η συνολικη παραγωγή ΦΒ με 2115/5 ώστε η μέγιστη παραγωγή ΦΒ να έιναι ίση με το μέγιστο φορτίο (5MW).\
Στο σενάριο επίσης, υπάρχει γεννήτρια diesel 5MW η οποία λειτουργεί ως εφεδρεία σε περιόδους χαμηλής η και μηδενικής παραγωγής ενεργειας.

## Προσωμοίωση Δικτύου
### Παρούσα Κατάσταση 
Για την διαμόρφωση του δικτύου χρησιμοποιήσαμε την βιβλιοθήκη pypsa της python.\
Το δίκυο αποτελείται από: 
- Δυο ζυγούς, οι οποίοι είναι ο ζυγός αναφοράς ("Diesel") όπου παραγεται ενεργεια μεσω γεννήτριας diesel ("Diesel Gen") και ο ζυγός παραγωγής ("Solar")  όπου υπάρχει το ΦΒ πάρκο ("PV Park") και το φορτίο ("load 1")
- Μία γραμμή μεταφοράς ("Line") η οποία ενώνει τον ζυγό αναφοράς με τον ζυγό παραγωγής
- Το φορτίο όπου η απόκρισή του δίνεται απο τα παραπάνω δεδομενα "data_load"

Αξίζει να σημειωθεί ότι για τα capital costs η τιμή που τέθηκε είναι μηδενική, καθώς θεωρούμε ότι αυτο το σύστημα υπάρχει ήδη ενώ για τα marginal costs έχουν υπολογιστεί για το ΦΒ παρκο τα λειτουργικά έξοδα σε €/MWh και για την γεννητρια diesel τα λειτουργικα έξοδα και ο φόρος από τα CO2 emissions.

## Προσωμοίωση Power to Gas  
### Σενάριο 1 (P2G_case1.ipynb)
Το Power to Gas Σύστημα αποτελείται από:
- Τον ηλεκτρολύτη PEM ονομαστικής ισχύος P=2MW, απόδοσης=0.7, αρχικού κόστους 1000€/kWh, λειτουργικού κόστους 0.11€/kWh
- Την κυψελη καυσίμου ονομαστικής ισχύος P=1.2MW, απόδοσης=0.6, αρχικού κόστους 1000€/kWh, λειτουργικού κόστους 0.11€/kWh
- Το store υδρογόνου χωρητικότητας=15MWh, αρχικού κόστους 40000€/MWh, λειτουργικού κόστους 10€/ΜWh
- Τον ζυγό "Hydrogen" όπου συνδέονται όλα τα παραπάνω.

To pypsa δεν έχει στοιχεία όπως ηλεκτρολύτη και fuel cell και για τον λόγο αυτο υλοποιήθηκαν χρησιμοποιώντας "Links" τα οποία επιτρέπουν την σύνδεση ζυγών και την μετατροπή ενεργειας από μια μορφή σε μία άλλη.
Επίσης, επειδή η επένδυση είναι διαρκειας 25 ετών ενώ τα δεδομένα που έχουμε διαθέσιμα ειναι περιπου 5 ετών κάναμε μια αναγωγή και τα capital costs διαιρέθηκαν με 5.


## Προσωμοίωση Power to Gas 2030
### Σενάριο 2 (P2G_case2.ipynb)
Το 2030 θεωρείται ότι τα αρχικά κόστη των στοιχείων του Power to Gas θα έχουν μειωθει έως και 70%.\
Για τον λόγο αυτό προβλεπεται ότι θα ισχύουν τα παρακατω:
- O ηλεκτρολύτης PEM ονομαστικής ισχύος P=2MW, απόδοσης=0.7, αρχικού κόστους 350€/kWh, λειτουργικού κόστους 0.03€/kWh
- Η κυψελη καυσίμου ονομαστικής ισχύος P=1.2MW, απόδοσης=0.6, αρχικού κόστους 350€/kWh, λειτουργικού κόστους 0.03€/kWh
- Το store υδρογόνου χωρητικότητας=15MWh, αρχικού κόστους 30000€/MWh, λειτουργικού κόστους 10€/ΜWh
- Το λειτουργικό κόστος της γεννήτριας diesel θα έχει αυξηθεί στα 500€/MWh
